```python
import pandas as pd 
df = pd.read_csv('diabetes.csv')
df.head() 
```

```python
df.info()
```

```python
correlations = df.corr(method='pearson')
correlations
```

```python
import matplotlib.pyplot as plt 

fig = plt.figure(figsize=(11,11))
ax = fig.add_subplot(1,1,1) 
cax = ax.matshow(correlations, vmin=-1, vmax=1) 
fig.colorbar(cax) 

ticks = [0,1,2,3,4,5,6,7,8]
ax.set_xticks(ticks)
ax.set_yticks(ticks) 

ax.set_xticklabels(["Pregnancies","Glucose","BP","SkinThickness","Insulin","BMI","DPF","Age","Outcome"])
ax.set_yticklabels(["Pregnancies","Glucose","BP","SkinThickness","Insulin","BMI","DPF","Age","Outcome"])
plt.show()
```

```python
xpoint = df['BMI']
ypoint = df['BloodPressure']
plt.scatter(xpoint, ypoint) 
plt.xlabel('BMI')
plt.ylabel('BloodPreassure') 
plt.show()
```

```python
import seaborn as sns 
sns.lineplot(data=df, x = 'Age', y='BloodPressure')
```
